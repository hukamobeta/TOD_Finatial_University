2. Проанализируйте временной ряд максимальной цены акций на предмет выбросов.

````python
mean = df['close'].mean()
std = df['close'].std()
print('Mean:', mean)
print('Standard deviation:', std)
lower_bound = mean - 2.2 * std
upper_bound = mean + 1 * std
print('Lower bound:', lower_bound)
print('Upper bound:', upper_bound)

outliers = df[(df['close'] < lower_bound) | (df['close'] > upper_bound)]['close']
print('Outliers:', outliers)
#outliers.index = [datetime.datetime(int(outliers.index[i].split('-')[0]), 
                                    #int(outliers.index[i].split('-')[1]), 
                                    #int(outliers.index[i].split('-')[2])) for i in range(len(outliers))]

df['close'].plot(figsize=(12, 6))
plt.plot(outliers.index, outliers.values, 'ro')
plt.xlabel('Date')
plt.ylabel('High Price')
plt.title('Apple Stock Prices')
plt.show()
````
